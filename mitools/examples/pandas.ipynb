```python
import mitools as mt
import pandas as pd
```

```python
df = pd.DataFrame({
                ("A", "one", 'a'): [1, 2, 3],
                ("A", "two", 'b'): [4, 5, 6],
                ("B", "three", 'a'): [7, 8, 9],
            })
df.columns = pd.MultiIndex.from_tuples(df.columns)
df.index = pd.MultiIndex.from_tuples([(1, 2, 3), (1, 2, 1), (2, 1, 3)], names=['x', 'y', 'z'])

df
```

```python
df_single = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6], "C": [7, 8, 9]})
arrays = [["A", "A", "B"], ["X", "Y", "Z"]]
columns = pd.MultiIndex.from_arrays(arrays, names=["upper", "lower"])
df_multi = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]], columns=columns, index=columns)
df_multi
```

```python
mt.ec.select_columns(df_multi, ["A"], level='upper') + mt.ec.select_columns(df_multi, ["B"], level='upper').values
```
